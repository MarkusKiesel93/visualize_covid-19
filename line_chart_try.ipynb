```python
import numpy as np
import pandas as pd

from datetime import date, timedelta

from bokeh.layouts import layout, column, row
from bokeh.models import ColumnDataSource, GeoJSONDataSource, Slider, Button
from bokeh.plotting import figure
from bokeh.io import show, output_notebook, curdoc

from data_handler import DataHandler

output_notebook()
```

```python
dh = DataHandler()
covid_data = dh.data[dh.date_range[1]]
geo_data = dh.geo_data
```

```python
TOOLS = 'hover,tap,pan'

source = ColumnDataSource(covid_data)
cases = figure(plot_width=800, 
               plot_height=500, 
               tools=TOOLS,
               toolbar_location='right', 
               x_axis_type="datetime", 
               background_fill_color="#efefef")

cases.hover.tooltips = [
    ('country', '@countriesAndTerritories'),
]

cases.line('date', 'cases', source=source)

show(cases)
```
